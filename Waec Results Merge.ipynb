````python
import pandas as pd # The only external library we will be using here.
import datetime # will be used later for formating date
````

````python
waec_data = pd.read_excel('.\\Raw Data\\waec 2016 to 2018.xlsx', sheet_name=None) #read all sheets
waec_data.keys() #check all the sheet names
````

````python
to_delete = {'Sheet1'} #unwanted sheets
for k in to_delete:
    waec_data.pop(k, None)
````

````python
# Before proceed lets look at how one the sheets looks like,
waec_data['private 2016'].transpose()
````

### Anticpated structure
The anticipated structure is something like:

---
> **YEAR | SCHOOL_TYPE | METRICS | GENDER | ABIA | ABUJA | ADAMAWA | *LOCATIONS...*** 
---

Thus, based on the look of the data, we wil have to carry out multple operations which we shall go through step by step here.   
Lets take one of the sheets through the steps before applying the same procedure to all the sheets.

````python
# First, transpose the sheet.
waec_sample = waec_data['private 2016'].transpose()
waec_sample.head(5)
````

````python
# The row names(index) look unfortunate, we can solve this by giving it new numbers as names.
waec_sample.reset_index(inplace=True, drop=True)
# The first two columns are meaningless
waec_sample=waec_sample[waec_sample.columns[3:]]
waec_sample
````

````python
# Now Lets fix the column names
waec_sample.columns = waec_sample.iloc[0] #setting the column names
# Deletin the first row
waec_sample=waec_sample[1:]
# setting the Gender and Metrics Column Name
waec_sample.rename(columns={waec_sample.columns[0]: 'METRICS',waec_sample.columns[1]:'GENDER'}, inplace=True)
waec_sample.head(5)
````

````python
# Again, for my own end use I don't need the Total rows on the gender column
# as I can always get them back simply summing up the male and female values.
waec_sample = waec_sample.query("GENDER != 'TOTAL'")
waec_sample.head(10) # view the first 5 lines
````


1. All the NaN values could be replaced by the preceeding metric value such that, for example, We have 
>TOTAL NUMBER SAT | MALE  
>TOTAL NUMBER SAT | FEMALE
2. Again, for our case, we can safely drop the last two rows  
Since they can be gotten back by simply doing  
> ```5 CREDITS & ABOVE INCLUDING MATHEMATICS & ENGLISH``` / ```TOTAL NUMBER SAT```

````python
# Pandas forward-fill, pd.ffill(), helps to, by default, replace NaN values with
# the previous value in the row,hence it saves us from some acrobatics.
waec_sample.METRICS.ffill(inplace=True)
#Forgetting the last two rows
waec_sample = waec_sample[:-2]
waec_sample
````

Good, we have two columns Left, Year and School type
The two column values will be gotten from the sheet as the sheet names seems to follow the same pattern  
>```private 2016``` implies school type = Private, Year = 2016.

Thus, lets proceed by generating the two columns from the sheet names

````python
# let say the sheet name is "private 2016"
# We will get this programatically when looping across the sheets later
sheet_name = "private 2016"
school_type, year = sheet_name.split()

# As with several other operation there are many ways of achieving this, but
# I choose to the df.insert method to save me time of rearranging the columns
waec_sample.insert(loc=0, column="YEAR", value=year) # Insert Year in the first column
waec_sample.insert(loc=1, column="SCHOOL_TYPE", value=school_type.upper()) # School in second column
waec_sample
````

````python
waec_sample
````

-----
Now, lets combine all the codes and apply them to all the sheets  
New lines of codes will be placed within ```#---------``` and ```#--------```


````python

full_waec_data = pd.DataFrame() # where the full waec data will be stored

for name, sheet in waec_data.items():    # Passing looping across the sheets for processing
    # First, transpose the sheet.
    waec_sample = sheet.copy().transpose(); #getting copy of the real sheet
    #waec_sample.head(5)
    # The row names(index) look unfortunate, we can solve this by giving it new numbers as names.
    waec_sample.reset_index(inplace=True, drop=True)
    # The first two columns are meaningless
    waec_sample=waec_sample[waec_sample.columns[2:]]
    #waec_sample

    # Now Lets fix the column names
    waec_sample.columns = waec_sample.iloc[0] #setting the column names
    # Deleting the first row
    waec_sample=waec_sample[1:]
    
    # When ruunning loops of dataframes, I always try to add this extra two lines of codes
    # which takes care of three common scenes.
    
    #-----------------------------------------------------------------
    # this one replace all empty string with NaN which are easier to handle
    waec_sample.replace(' ',pd.np.nan, inplace=True)
    # this eliminate all empty columns, that might have been read into the dataframe
    waec_sample.dropna(axis=1,how='all',inplace=True)
    # this drops all empty rows that might have read into the data frame
    waec_sample.dropna(axis=0,how='all',inplace=True)
    #------------------------------------------------------------------
    
    # setting the Gender and Metrics Column Name
    waec_sample.rename(columns={waec_sample.columns[0]: 'METRICS',waec_sample.columns[1]:'GENDER'}, inplace=True)
    #waec_sample.head(5)

    # Again, for my own end use I don't need the Total rows on the gender column
    # as I can always get them back simply summing up the male and female values.
    waec_sample = waec_sample.query("GENDER != 'TOTAL'")
    #waec_sample.head(10) # view the first 5 lines

    '''
    1.  All the NaN values could be replaced by the preceeding metric value such that, for example, We have 
        TOTAL NUMBER SAT | MALE  
        TOTAL NUMBER SAT | FEMALE
    2.  Again, for our case, we can safely drop the last two rows  
        Since they can be gotten back by simply doing  
        5 CREDITS & ABOVE INCLUDING MATHEMATICS & ENGLISH / TOTAL NUMBER SAT
    '''
    # Pandas forward-fill, pd.ffill(), helps to, by default, replace NaN values with
    # the previous value in the row,hence it saves us from some acrobatics.
    waec_sample.METRICS.ffill(inplace=True)
    #Forgetting the last two rows
    waec_sample = waec_sample[:-2]
    #waec_sample

    '''
    Good, we have two columns Left, Year and School type
    The two column values will be gotten from the sheet as the sheet names seems 
    to follow the same pattern  
    "private 2016" implies: school type = Private, Year = 2016.

    Thus, lets proceed by generating the two columns from the sheet names
    '''

    # The sheet name is accessible within this loop as name.
    school_type, year = name.split()

    # As with several other operations there are many ways of achieving this, but
    # I choose to the df.insert method to save me time of rearranging the columns
    waec_sample.insert(loc=0, column="YEAR", value=year) # Insert Year in the first column
    waec_sample.insert(loc=1, column="SCHOOL_TYPE", value=school_type.upper()) # School in second column
    
    
    #------------------------------------------
    # We will want to append our frames onto each other once they have been successfully processed
    # The sort=False argument ensures that the dataframe is not rearranged alphabetically
    full_waec_data = full_waec_data.append(waec_sample, sort=False)
    #------------------------------------------
    
full_waec_data
````

## Looks great!!! 
Lets take care of the messed-up index and save it. Then we will call it a day

````python
full_waec_data.reset_index(drop=True, inplace=True) # This re-numbers the index from 0
full_waec_data.to_csv(".\\output\\Waec_2016_2018.csv") # Saving as Waec_2016_2018.csv file
````
